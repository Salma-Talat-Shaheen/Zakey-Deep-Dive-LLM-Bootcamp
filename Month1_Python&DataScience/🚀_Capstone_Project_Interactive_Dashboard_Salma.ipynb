````python
%%writefile app.py
import streamlit as st
import pandas as pd
import plotly.express as px

# --- Page Configuration ---
st.set_page_config(
    page_title="Global Temperature Dashboard",
    page_icon="üåç",
    layout="wide"
)

# --- General CSS for Backgrounds & Sections ---
st.markdown("""
<style>
/* Overall page background */
body, .stApp {
    background-color: #fff8dc; /* Light beige/yellow */
}

/* Section containers */
.section {
    padding: 15px;
    border-radius: 10px;
    margin-bottom: 20px;
    color: #2c3e50;
}

/* Gradients for sections */
.section-gradient-blue {
    background: linear-gradient(to right, #d6eaf8, #85c1e9);
}

.section-gradient-light {
    background: linear-gradient(to right, #fef9e7, #f9e79f);
}

h1, h2, h3, h4, h5 {
    color: #2c3e50;
}
</style>
""", unsafe_allow_html=True)

# --- Load Data ---
@st.cache_data
def load_data():
    url = "https://raw.githubusercontent.com/Steven-Alvarado/Global-Temperature-Analysis/refs/heads/main/GlobalTemperatures.csv"
    df = pd.read_csv(url)
    df['dt'] = pd.to_datetime(df['dt'])
    df['Year'] = df['dt'].dt.year
    df.dropna(subset=['LandAverageTemperature'], inplace=True)
    return df

df = load_data()
yearly_temp_full = df.groupby('Year')['LandAverageTemperature'].mean().reset_index()

# --- Title Section ---
st.markdown("""
<div class="section section-gradient-light">
    <h1 style="text-align:center;">Global Temperature Trends Dashboard</h1>
    <p style="text-align:center;">Explore global land temperatures over time with interactive charts and filters.</p>
</div>
""", unsafe_allow_html=True)

# --- Top Metrics Section ---
st.markdown("""
<div class="section section-gradient-blue">
    <h4 style="text-align:center;">Key Metrics</h4>
</div>
""", unsafe_allow_html=True)

# --- Centered Key Metrics ---
st.markdown("""
<div style="display: flex; justify-content: center; text-align: center; gap: 50px;">
    <div>
        <h3>Records</h3>
        <p style="font-size:20px; font-weight:bold;">{}</p>
    </div>
    <div>
        <h3>Years Covered</h3>
        <p style="font-size:20px; font-weight:bold;">{} - {}</p>
    </div>
    <div>
        <h3>Average Temp</h3>
        <p style="font-size:20px; font-weight:bold;">{:.2f} ¬∞C</p>
    </div>
</div>
""".format(len(df), df['Year'].min(), df['Year'].max(), yearly_temp_full['LandAverageTemperature'].mean()), unsafe_allow_html=True)


# --- Filters & Chart Section ---
st.markdown("---")
st.subheader("Interactive Exploration")

left_col, right_col = st.columns([1, 4])

with left_col:
    st.markdown('<div class="section section-gradient-light"><h5>Filters</h5></div>', unsafe_allow_html=True)
    year_range = st.select_slider(
        "Select Year Range",
        options=sorted(df['Year'].unique()),
        value=(1900, 2015)
    )
    chart_type = st.radio(
        "Chart Type",
        ["Line", "Bar", "Area"],
        index=0
    )

with right_col:
    filtered_df = df[(df['Year'] >= year_range[0]) & (df['Year'] <= year_range[1])]
    yearly_temp = filtered_df.groupby('Year')['LandAverageTemperature'].mean().reset_index()

    st.markdown('<div class="section section-gradient-blue"><h4>Temperature Trends</h4></div>', unsafe_allow_html=True)

    if chart_type == "Line":
        fig = px.line(yearly_temp, x='Year', y='LandAverageTemperature',
                      labels={'LandAverageTemperature': 'Avg Temp (¬∞C)'})
        fig.add_scatter(
            x=yearly_temp['Year'],
            y=yearly_temp['LandAverageTemperature'].rolling(window=10).mean(),
            mode='lines',
            name='10-Year Moving Avg',
            line=dict(color="orange", width=3)
        )
    elif chart_type == "Bar":
        fig = px.bar(yearly_temp, x='Year', y='LandAverageTemperature',
                     labels={'LandAverageTemperature': 'Avg Temp (¬∞C)'})
    else:
        fig = px.area(yearly_temp, x='Year', y='LandAverageTemperature',
                      labels={'LandAverageTemperature': 'Avg Temp (¬∞C)'})

    st.plotly_chart(fig, use_container_width=True)

# --- Dataset Info Section ---
st.markdown("---")
st.subheader("Dataset Information & Description")
st.markdown("""
<div class="section section-gradient-light">
<b>Source:</b> Berkeley Earth Surface Temperature Project<br>
This dataset contains monthly and annual global land temperature averages with uncertainty measures.<br>
Coverage: 1750 ‚Äì 2015<br>
Variables: Average, Max, Min temperatures, and combined Land & Ocean<br>
Frequency: Monthly records<br>
For more info: <a href='http://berkeleyearth.org/data/' target='_blank'>Berkeley Earth</a>
</div>
""", unsafe_allow_html=True)

# --- Dataset Descriptive Statistics ---
st.subheader("Descriptive Statistics")
st.markdown("""
<div class="section section-gradient-blue">
This section shows a quick summary of the dataset using pandas `describe()`.
</div>
""", unsafe_allow_html=True)

st.dataframe(df.describe())

# --- Raw Data ---
with st.expander("Show Filtered Data"):
    st.dataframe(filtered_df)
````
