#### **1. Get User Input**

````python
# Ask the user to enter numbers
input_str = input("Enter numbers separated by commas: ")

# Convert the input string into a list
number_list = [int(num.strip()) for num in input_str.split(',')]

print("Numbers entered:", number_list)
````

#### **2. Store Frequencies in a Dictionary**

````python
# Create a dictionary
frequency_dict = {}

#  count frequencies
for num in number_list:
    if num in frequency_dict:
        frequency_dict[num] += 1
    else:
        frequency_dict[num] = 1

print("Frequency Dictionary:", frequency_dict)
````

#### **3. Define a Function to Process the Numbers**

````python
def calculate_stats(numbers):
    total = sum(numbers)
    avg = round(total / len(numbers), 3)
    maximum = max(numbers)
    minimum = min(numbers)
    return total, avg, maximum, minimum
````

#### **4. Call the Function and Get Results**

````python
sum_result, avg_result, max_result, min_result = calculate_stats(number_list)
````

#### **5. Display Results in a Markdown Table**

````python
# Print results as a Markdown table
table= f"""
| Metric      | Value |
|-------------|-------|
| Sum         | {sum_result}    |
| Average     | {avg_result}   |
| Maximum     | {max_result}     |
| Minimum     | {min_result}     |
"""
display(Markdown(table))
````

#### **6. Bonus**

````python
# 1. Sorting the numbers
sorted_numbers = sorted(number_list)
print(" Sorted Numbers (Ascending):", sorted_numbers)
````

````python
# 2. Most frequent number
most_frequent = max(frequency_dict, key=frequency_dict.get)
print(" Most Frequent Number:", most_frequent, "‚Üí appeared", frequency_dict[most_frequent], "times")
````

````python
# 3. Plotting a bar chart of frequencies
import matplotlib.pyplot as plt

plt.bar(frequency_dict.keys(), frequency_dict.values(), color="red", edgecolor="black")
plt.xlabel("Numbers")
plt.ylabel("Frequency")
plt.title("Number Frequency Chart")
plt.show()
````
